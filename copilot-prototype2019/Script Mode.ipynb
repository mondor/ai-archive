```python
from sagemaker.tensorflow import TensorFlow

model = TensorFlow(entry_point='multilabels_model_local.py',
                             role=role,
                             framework_version='1.12.0',
                             train_instance_count=1,
                             py_version='py3',
                             script_mode=True,
                             train_instance_type='ml.m4.xlarge',
                             hyperparameters={
                                 'input_dim': 150, 
                                 'n_classes': 2,
                                 'num_epochs': 2,
                                 'batch_size': 256,
                                 'embed_dim': 300,
                                 'pretrain_embedding': 's3://sagemaker-us-east-1-953585160895/pretrain-embedding/wiki.simple.vec'
                             })

model.fit('s3://sagemaker-us-east-1-953585160895/data/5c46a5bf830c3')
```

```python
training_job_name = model.latest_training_job.name
job = sagemaker_session.sagemaker_client.describe_training_job(TrainingJobName=training_job_name)
```
