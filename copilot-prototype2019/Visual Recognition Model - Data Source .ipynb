```python
import pandas as pd
import os
import numpy as np

df = pd.read_csv('input/data/training/image_data.txt', header=None, names=["id", "source", "image", "original_url", "published", "disabled"])

data = df.values

np.random.seed(42)
np.random.shuffle(data)


offset = int(data.shape[0] * 0.8)

train_data, test_data = data[:offset, :], data[offset:, :]



image_root_folder = 'input/data/training/images'
image_folder_test = os.path.join(image_root_folder, 'test')
image_folder_train = os.path.join(image_root_folder, 'train')


if not os.path.isdir(image_folder_test):
    os.mkdir(image_folder_test)
    
if not os.path.isdir(image_folder_train):
    os.mkdir(image_folder_train)
    
    
    
```

```python
import re
def ig_image(url):
    shortcode_search = re.search('/p/([^/]*)', url, re.IGNORECASE)
    if shortcode_search:
        return 'https://www.instagram.com/p/{}/media?size=m'.format(shortcode_search.group(1))
    return ''
```

```python
import urllib
import traceback


def save_image(folder, _id, url):
    if not os.path.isdir(folder):
        os.mkdir(folder)
    
    with urllib.request.urlopen(url, timeout=2) as response:
        info = response.info()
        mime_type = info.get_content_type()
        
        ext = ''
        if mime_type == 'image/png':
            ext = 'png'
        elif mime_type == 'image/jpeg':
            ext = 'jpg'
            
        local_file = os.path.join(folder, _id + '.' + ext)        
            
        assert(ext)
        
        f = open(local_file, 'wb')
        f.write(response.read())
        f.close()
        
        print(local_file)
                
        
        

def save_images(dest_folder, data):
    for row in data:
        _id = row[0]
        source = row[1]
        url = row[2]
        original_url = row[3]
        is_published = row[4]
        
        
        if is_published:
            label_folder = os.path.join(dest_folder, 'published')
        else:
            label_folder = os.path.join(dest_folder, 'queued')

        try:
            save_image(label_folder, _id, url)
        except Exception as e:
            print(str(e))            
            if source == "instagram" and original_url:                
                try:
                    url = ig_image(original_url)                
                    print('try ' + url)
                    save_image(label_folder, _id, url)
                except Exception as e:
                    print(str(e))


            
        


```

```python
save_images(image_folder_train, train_data)
```

```python
save_images(image_folder_test, test_data)
```
