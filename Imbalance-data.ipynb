```python
from textblob import TextBlob
from textblob.translate import NotTranslated
import random


```

```python
def robust_decode(bs):
    '''Takes a byte string as param and convert it into a unicode one.
First tries UTF8, and fallback to Latin1 if it fails'''
    cr = None
    try:
        cr = bs.decode('utf8')
    except UnicodeDecodeError:
        cr = bs.decode('latin1')
    return cr

source_text = u"the pot of gold was out of reach at this stage"

blob = TextBlob(source_text)
sl = blob.detect_language()
inbetween_text = robust_decode(str(blob.translate(to='zh-CN')))

print(inbetween_text)
blob2 = TextBlob(inbetween_text)
final_text = str(blob2.translate(to='en'))

print(final_text)


```
